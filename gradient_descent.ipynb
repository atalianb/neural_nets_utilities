```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
```

```python
func = lambda th: np.sin(1/2*th[0]**2 - 1/4*th[1]**2 +3)*np.cos(2*th[0]+1-np.exp(th[1]))
```

```python
func([1,1])
```

```python
X = np.linspace(-2,2,100)
Y = np.linspace(-2,2,100)
Z = np.zeros((100,100))
```

```python
for ix,x in enumerate(X):
    for iy, y in enumerate(Y):
        Z[iy,ix]=func([x,y])
```

```python
plt.contour(X,Y,Z,100)
plt.colorbar()
```

```python
plt.contourf(X,Y,Z,100)
plt.colorbar()

#dos valores aleatorios de -2 a 2
theta = np.random.rand(2)*4-2
_theta = np.copy(theta)
h=0.001
lrate = 0.0001
gradient = np.zeros(2)

plt.plot(theta[0],theta[1],"o",color='r')

for _ in range(10000):
    for it, th in enumerate(theta):
        _theta=np.copy(theta)
        _theta[it]+=h
        der = (func(_theta)-func(theta))/h
        gradient[it] = der
    
    theta = theta - lrate * gradient
    #print(func(theta))
    if _%50==0:
        plt.plot(theta[0],theta[1],".",color='black')


plt.plot(theta[0],theta[1],"o",color='pink')
plt.show
```
