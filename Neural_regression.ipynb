```python
import keras
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

Cons√≠derese el modelo lineal $Y=1.5X+2$ :

```python
X=np.linspace(0,5,50) 
Y=1.5*X+2
plt.scatter(X,Y,label='y=1.5x+2',marker='.')
plt.title('Linear model')
plt.legend()
plt.show()
```

Para simular datos observacionales, le agregamos un ruido gaussiano:

```python
noise=np.random.normal(0,1,50) 
Y=Y+noise 
plt.scatter(X,Y,label='$y=1.5x+c+N(\mu=0,\sigma=1)$') 
plt.legend() 
plt.title('Model with gaussian noise')
plt.show()
```
